```python
with open('data_scientist_exercise_file.fq') as f:
    lines=f.readlines()

data=[item[:-1] for item in lines[1::4]]
```

```python
def Find_Reverse(DNA):
    dic = {'A':'T', 'T':'A', 'G':'C', 'C':'G'}
    barcode1 = DNA[0:6]
    barcode2 = ''
    for i in barcode1:
        barcode2 += dic[i]
    return barcode1, barcode2[::-1]
```

```python
Find_Reverse('AGTGGCCCTGGAATTGTCTCTGTTCAG')
```

```python
columns = {'Barcode1', 'NumberB1', 'Barcode2', 'NumberB2', 'DNA'}
df = pd.DataFrame(index=range(len(data)), columns=columns)
df.NumberB1 = df.fillna(0)
df.NumberB2 = df.fillna(0)

counter = 0

for DNA in data:
    bar1, bar2 = Find_Reverse(DNA)
 
    if sum(df.Barcode1 == bar1):
        df.NumberB1[df.Barcode1 == bar1] += 1
    elif sum(df.Barcode2 == bar1):
        df.NumberB2[df.Barcode2 == bar1] += 1
    else:
        df.DNA[counter] = DNA
        df.Barcode1[counter] = bar1
        df.Barcode2[counter] = bar2
        df.NumberB1[counter] += 1
        
#     if DNA == 'ATGACAGCCTGCTGCGTGTTGACTGCA':
#     if bar1 == 'CCACCT':
#         print('a')
#         print(df.NumberB1[counter])
        
    counter +=1
```

```python
df = df.dropna()
df  
```

```python
# re order based on minimum coverage
for i in range(len(data)):
    try:
        if df.NumberB1[i] > df.NumberB2[i]:
            df.NumberB2[i], df.NumberB1[i] = df.NumberB1[i], df.NumberB2[i]
            df.Barcode2[i], df.Barcode1[i] = df.Barcode1[i], df.Barcode2[i]
    except:
        continue
```
